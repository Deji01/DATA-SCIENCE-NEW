```python
CovidDeaths = df[['continent', 'location', 'date', 'total_cases', 'new_cases','total_deaths','new_deaths','population','total_tests', 'total_vaccinations',
       'people_vaccinated',]] 
```

```python
CovidDeaths.info()
```

```python
CovidDeaths = CovidDeaths[CovidDeaths['continent'].notnull()]
```

```python
CovidDeaths.describe()
```

```python
correlation_matrix = CovidDeaths.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title("Correlation matrix for Numeric Features")
plt.xlabel("Covid death features")
plt.ylabel("Covid death features")

plt.show()
```

```python
plt.scatter(x=CovidDeaths['total_cases'], y=CovidDeaths['total_deaths'], alpha=0.5)
plt.title('Total Cases vs Total Deaths')
plt.xlabel('Total Cases')
plt.ylabel('Total Deaths')
plt.show()
```

```python
sns.regplot(x='total_cases', y='total_deaths', data=CovidDeaths,line_kws={'color':'black'})
```

```python
plt.scatter(x=CovidDeaths['new_cases'], y=CovidDeaths['new_deaths'], alpha=0.5)
plt.title('New Cases vs New Deaths')
plt.xlabel('New Cases')
plt.ylabel('New Deaths')
plt.show()
```

```python
sns.regplot(x='new_cases', y='new_deaths', data=CovidDeaths,line_kws={'color':'black'})
```

```python
# Total Cases vs Total Deaths
# Shows likelihood of dying if you contact covid in a country

TotalDeaths = CovidDeaths['total_deaths']
TotalCases = CovidDeaths['total_cases']
```

```python
CovidDeaths['fatality_rate'] = round((TotalDeaths/ TotalCases * 100), 2)
```

```python
#  Total Cases vs Population
# Shows what percentage of population infected with Covid

Population = CovidDeaths['population']
```

```python
CovidDeaths['pct_pop_infected'] = round((TotalCases / Population * 100), 2)
```

```python
#  Total Deaths vs Population
# Shows what percentage of population who died after contacting Covid

CovidDeaths['mortality_rate'] = round((TotalDeaths / Population * 100), 2)
```

```python
CovidDeaths.tail()
```

```python
CovidDeaths.groupby(['location'])['fatality_rate'].mean().sort_values(ascending=False).head(6)
```
